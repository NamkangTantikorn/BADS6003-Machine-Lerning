```python
# Quartiles
print('1 Quartile: ', df_train['age'].quantile(q = 0.25))
print('2 Quartile: ', df_train['age'].quantile(q = 0.50))
print('3 Quartile: ', df_train['age'].quantile(q = 0.75))
print('4 Quartile: ', df_train['age'].quantile(q = 1.00))

#Calculate the outliers:
  # Interquartile range, IQR = Q3 - Q1
  # lower 1.5*IQR whisker = Q1 - 1.5 * IQR 
  # Upper 1.5*IQR whisker = Q3 + 1.5 * IQR
    
print('Ages above: ', df_train['age'].quantile(q = 0.75) + 
                      1.5*(df_train['age'].quantile(q = 0.75) - df_train['age'].quantile(q = 0.25)), 'are outliers')
print('Ages below: ', df_train['age'].quantile(q = 0.75) - 
                      1.5*(df_train['age'].quantile(q = 0.75) - df_train['age'].quantile(q = 0.25)), 'are outliers')
```

```python
from scipy.stats import ttest_ind
data1 = df_train.age[df_train.target==1]
data2 = df_train.age[df_train.target==0]
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably the same distribution')
else:
    print('Probably different distributions')
```
